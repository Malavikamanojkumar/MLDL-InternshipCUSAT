```python
import numpy as np
import pandas as pd

# --- Step 1: Define Vocabulary and Tokenization ---
# Our sample vocabulary
vocabulary = ['the', 'cat', 'sat', 'on', 'the', 'mat', 'quickly', 'jumped']

# Create a mapping from word to unique integer ID
word_to_id = {word: i for i, word in enumerate(sorted(list(set(vocabulary))))}
id_to_word = {i: word for word, i in word_to_id.items()}

print("--- Step 1: Vocabulary and Token IDs ---")
print("Word to ID mapping:", word_to_id)
print("ID to Word mapping:", id_to_word)

# Example word and its token ID
word = 'cat'
token_id = word_to_id.get(word, -1) # Use .get() to handle words not in vocab
print(f"\nWord '{word}' corresponds to Token ID: {token_id}")
```

```python
# --- Step 2: From Token ID to a d-dimensional Feature Vector (Embedding) ---
# Define the embedding dimension (d)
embedding_dim = 5

print("\n--- Step 2: Individual Embeddings (d-dimensional Feature Vectors) ---")

# Store individual embeddings in a dictionary
individual_embeddings = {}
for word, token_id in word_to_id.items():
    # Generate a random vector for each token ID
    # np.random.rand() generates values between 0 and 1. We scale and shift them.
    embedding = (np.random.rand(embedding_dim) * 2 - 1) # Values between -1 and 1
    individual_embeddings[token_id] = embedding
    print(f"Token ID {token_id} ('{word}'): {embedding.round(3)}")

# Example: Embedding for 'cat' (Token ID 1)
if token_id != -1:
    cat_embedding = individual_embeddings.get(token_id)
    print(f"\nEmbedding for '{word}' (Token ID {token_id}): {cat_embedding.round(3)}")
else:
    print(f"Word '{word}' not found in vocabulary.")
```

```python
# --- Step 3: Stacking to Form the Embedding Matrix ---
# Get the size of the vocabulary (V)
vocab_size = len(word_to_id)

# Create an empty matrix to hold all embeddings
# The shape will be (vocab_size, embedding_dim)
embedding_matrix = np.zeros((vocab_size, embedding_dim))

# Fill the embedding matrix using the token IDs as indices
print("\n--- Step 3: Constructing the Embedding Matrix ---")
for token_id, embedding_vector in individual_embeddings.items():
    embedding_matrix[token_id] = embedding_vector

print(f"Embedding Matrix Shape: {embedding_matrix.shape} (Vocabulary Size x Embedding Dimension)\n")

# Display the embedding matrix using a Pandas DataFrame for better readability
# The index will be the Token IDs, and columns can be 'Dim_0', 'Dim_1', etc.
embedding_df = pd.DataFrame(embedding_matrix,
                            index=[id_to_word[i] for i in range(vocab_size)],
                            columns=[f'Dim_{i}' for i in range(embedding_dim)])

print("Embedding Matrix:")
display(embedding_df.round(3))
```

```python

if 'word_to_id' not in locals():
    vocabulary = ['the', 'cat', 'sat', 'on', 'the', 'mat', 'quickly', 'jumped']
    word_to_id = {word: i for i, word in enumerate(sorted(list(set(vocabulary))))}
    id_to_word = {i: word for word, i in word_to_id.items()}

if 'embedding_dim' not in locals():
    embedding_dim = 5

if 'individual_embeddings' not in locals():
    individual_embeddings = {}
    for word, token_id in word_to_id.items():
        embedding = (np.random.rand(embedding_dim) * 2 - 1)
        individual_embeddings[token_id] = embedding

print("--- Step 4: Implementing Positional Encoding ---")

def positional_encoding(position, d_model):
    """
    Calculates the positional encoding for a given position and embedding dimension.
    PE(pos, 2i) = sin(pos / 10000^(2i/d_model))
    PE(pos, 2i+1) = cos(pos / 10000^(2i/d_model))
    """
    pe = np.zeros(d_model)
    for i in range(d_model):
        if i % 2 == 0: # Even dimensions
            pe[i] = np.sin(position / (10000 ** (i / d_model)))
        else: # Odd dimensions
            pe[i] = np.cos(position / (10000 ** ((i - 1) / d_model)))
    return pe

# --- Step 5: Apply Positional Encoding to a Sequence ---
# Let's consider a sample sentence (sequence of words)
sentence = ['the', 'cat', 'sat', 'on', 'the', 'mat']

# Convert sentence words to token IDs
sentence_token_ids = [word_to_id[word] for word in sentence if word in word_to_id]

print(f"\nSample Sentence: {sentence}")
print(f"Token IDs for sentence: {sentence_token_ids}")

# Retrieve word embeddings for the sentence
word_embeddings_sequence = np.array([individual_embeddings[token_id] for token_id in sentence_token_ids])

# Generate positional encodings for each word in the sentence
positional_encodings_sequence = np.array([
    positional_encoding(pos, embedding_dim) for pos in range(len(sentence_token_ids))
])

print(f"\nShape of Word Embeddings for sequence: {word_embeddings_sequence.shape}")
print(f"Shape of Positional Encodings for sequence: {positional_encodings_sequence.shape}")

print("\nPositional Encodings for the sentence:")
positional_encoding_df = pd.DataFrame(positional_encodings_sequence,
                           index=[f'{word} (pos {i})' for i, word in enumerate(sentence)],
                           columns=[f'Dim_{i}' for i in range(embedding_dim)])
display(positional_encoding_df.round(3))

# --- Step 6: Combine Word Embeddings and Positional Encodings ---
# Add the positional encodings to the word embeddings
# This is a common way to inject position information into the embeddings
combined_embeddings = word_embeddings_sequence + positional_encodings_sequence

print("\nCombined Embeddings (Word Embedding + Positional Encoding) for the sentence:")
combined_df = pd.DataFrame(combined_embeddings,
                           index=[f'{word} (pos {i})' for i, word in enumerate(sentence)],
                           columns=[f'Dim_{i}' for i in range(embedding_dim)])
display(combined_df.round(3))

print("\n--- Explanation ---")
print("1. The `positional_encoding` function generates a vector where odd and even dimensions are filled with sine and cosine functions, respectively, based on the word's position and the embedding dimension.")
print("2. For a given sentence, we get the word embeddings for each token.")
print("3. We then generate a corresponding positional encoding for each position (0, 1, 2, ...).")
print("4. Finally, these positional encodings are simply added to their respective word embeddings. This sum is the new, position-aware embedding for each word in the sequence.")
```


1.  **Vocabulary and Token IDs**: We define a simple vocabulary and create a mapping from words to unique integer IDs. This represents the 'tokenization' step.
2.  **Individual Embeddings**: For each unique token ID, we generate a random `d`-dimensional numerical vector. In real-world applications, these vectors are learned through neural networks (e.g., Word2Vec, GloVe, BERT) to capture semantic relationships.
3.  **Embedding Matrix**: All these individual `d`-dimensional vectors are then stacked together, forming a `|V| x d` matrix. Each row in this matrix corresponds to the embedding of a specific word in our vocabulary.

This `embedding_matrix` is a fundamental component in many NLP models, allowing words to be represented as numerical data that machine learning algorithms can process.
